# ***LET'S SEE MODE, MEAN AND MEDIAN WITH AGE COLUMN MISSING VALUE***

## **INDEX OF MISSING VALUES IN AGE COLUMN WITH MEAN**

```python

from sklearn.impute import SimpleImputer
train = df.copy()

mean_imputer = SimpleImputer(strategy = 'mean')
train['Age'] = mean_imputer.fit_transform(train[['Age']])

train.iloc[age_missing_index]

```

## **INDEX OF MISSING VALUES IN AGE COLUMN WITH MEDIAN**

```python
train = df.copy()

median_imputer = SimpleImputer(strategy = 'median')
train['Age'] = median_imputer.fit_transform(train[['Age']] )

train.iloc[age_missing_index]
```

## **INDEX OF MISSING VALUES IN AGE COLUMN WITH MODE**

```python
train = df.copy()

mode_imputer = SimpleImputer(strategy = 'most_frequent')
train['Age']= mode_imputer.fit_transform(train[['Age']])

train.iloc[age_missing_index]
```

## **Category wise mean/median**

```python
train = df.copy()

train['Age'] = train.groupby('Title')['Age'].transform(lambda x: x.fillna(x.median()))
train.iloc[age_missing_index]
```

```python
train = df.copy()

train['Age'] = train.groupby('Title')['Age'].transform(lambda x: x.fillna(x.mean()))
train.iloc[age_missing_index]
```

# **K-Nearest Neighbor Imputation and Multivariate Feature Imputation With Age Column Missing Value.**

## **K-Nearest Neighbor Imputation**

```python
gender_mapper = {'male': 0, 'female': 1}
df['Sex'].replace(gender_mapper, inplace=True)
```

```python
df['Title'] = [0 if x in ['Mr.', 'Miss.', 'Mrs.'] else 1 for x in df['Title']]
df = df.rename(columns={'Title': 'Title_Unusual'})
```

```python
df
```

```python
from sklearn.impute  import KNNImputer

train_knn = df.copy()

knn_imputer = KNNImputer(n_neighbors=5, weights= 'uniform')

train_knn[['Age', 'Sex',	'Title_Unusual']] = knn_imputer.fit_transform(train_knn[['Age', 'Sex',	'Title_Unusual']])

train_knn.iloc[age_missing_index]
```

## **Multivariate Feature Imputation - Multivariate Imputation by Chained Equations (MICE)**

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

train_mice = df.copy()

mice_imputer = IterativeImputer()

train_mice[['Age', 'Sex',	'Title_Unusual']] = mice_imputer.fit_transform(train_mice[['Age', 'Sex',	'Title_Unusual']])

train_mice.iloc[age_missing_index]
```
